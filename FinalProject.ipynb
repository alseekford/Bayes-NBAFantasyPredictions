```python
data['10_game_average'] = data.groupby('player_id').fantasy_points.apply(lambda x: x.shift().rolling(10, min_periods=1).mean().fillna(x))
```

```python
data.avg_pts.describe()
```

```python
# create a list of our conditions
conditions = [
    (data['fantasy_points'] <= 0) & (data['avg_pts'] < 2),
    (data['fantasy_points'] > 0) & (data['fantasy_points'] < 3.3) & (data['avg_pts'] >= 2) & (data['avg_pts'] < 6.5),
    (data['fantasy_points'] >= 3.3) & (data['fantasy_points'] < 15.6) & (data['avg_pts'] >= 6.5) & (data['avg_pts'] < 16),
    (data['fantasy_points'] >= 15.6) & (data['fantasy_points'] < 27.5) & (data['avg_pts'] >= 16) & (data['avg_pts'] < 25), 
    (data['fantasy_points'] >= 27.5) & (data['fantasy_points'] < 100) & (data['avg_pts'] >= 25) & (data['avg_pts'] < 73)]

# create a list of the values we want to assign for each condition
values = [0, 1, 2, 3, 4]

# create a new column and use np.select to assign values to it using our lists as arguments
data['rank'] = np.select(conditions, values)
```
